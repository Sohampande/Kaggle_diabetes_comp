# Summary

A general overview is :

Strong feature skew and heavy tails exist in multiple numeric variables (e.g., physical activity, triglycerides, screen time).

Several features are predictive only through subtle distributional shifts, not extreme values (e.g., waist-to-hip ratio, blood pressure).

Categorical variables are highly imbalanced, with rare categories that can introduce noise or instability.

Many plots conditioned on diagnosis describe population composition, not risk, reinforcing the need for proper feature comparison and encoding rather than visual intuition.

The dataset includes variables with very different scales and statistical behavior, making naive scaling or encoding suboptimal.

Overall, the dataset is well-suited for prediction, but raw features should not be used directly without targeted preprocessing.

### Changes to be made to the dataset !!

1. **Drop non-informative or harmful features**

Remove id entirely. It contains no predictive signal and introduces noise. Including it can degrade performance, especially in linear or distance-based models.

2. **Transform heavily skewed numeric features**

Several features show long right tails and many outliers that dominate loss functions. Applying log transformations (log1p) to:
- physical_activity_minutes_per_week
- triglycerides
- ldl_cholesterol

This should reduce variance dominance, improve linear separability, stabilize gradient-based optimization

3. **Use robust or mixed scaling instead of global StandardScaler**

Avoid applying the same scaler to all numeric features. Recommended approach :
- Use StandardScaler for approximately normal features (age, BP, sleep)
- Use RobustScaler for skewed or outlier-heavy features (BMI, triglycerides, activity)
- Use ColumnTransformer to apply scalers selectively

4. **Preserve clinically meaningful “outliers”**

Many extreme values (e.g., high BMI, triglycerides, blood pressure) are not errors.
